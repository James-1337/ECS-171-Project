# Exploratory Data Analysis: Student Habits vs Exam Score

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("students.csv")


sns.set(style="whitegrid")


features_to_plot = {
    "exam_score": "Exam Score",
    "study_hours_per_day": "Study Hours Per Day",
    "social_media_hours": "Social Media Hours Per Day",
    "netflix_hours": "Netflix Hours Per Day",
    "attendance_percentage": "Attendance Percentage",
    "sleep_hours": "Sleep Hours",
    "exercise_frequency": "Exercise Frequency",
    "mental_health_rating": "Mental Health Rating"
}


for feature, label in features_to_plot.items():
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))


    sns.histplot(df[feature], kde=True, ax=axs[0])
    axs[0].set_title(f"Distribution of {label}")
    axs[0].set_xlabel(label)



    if feature != "exam_score":
        sns.scatterplot(x=df[feature], y=df["exam_score"], ax=axs[1])
        axs[1].set_title(f"{label} vs Exam Score")
        axs[1].set_xlabel(label)
        axs[1].set_ylabel("Exam Score")

    else:
        axs[1].axis('off')

    plt.tight_layout()
    plt.show()
```
